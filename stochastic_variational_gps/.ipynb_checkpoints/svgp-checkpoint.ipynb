# Stochastic Variational Gaussian Processes

SVGP is generally more robust and flexible because it allows you to learn the optimal inducing points through stochastic optimization, rather than pre-selecting them deterministically, as with the general inducing point methods.

```python
import torch
import gpytorch
from matplotlib import pyplot as plt
from torch.utils.data import TensorDataset, DataLoader
```

### Simple Latents

```python
torch.manual_seed(0)

# Generate 100,000 data points
N = 100000
X = torch.linspace(0, 10, N)
Y = torch.sin(X) + torch.randn(N) * 0.2  # Sinusoidal data with noise
```

```python
class SVGPModel(gpytorch.models.ApproximateGP):
    def __init__(self, inducing_points):
        # Variational distribution and strategy
        variational_distribution = gpytorch.variational.CholeskyVariationalDistribution(inducing_points.size(0))
        variational_strategy = gpytorch.variational.VariationalStrategy(
            self, inducing_points, variational_distribution, learn_inducing_locations=True
        )
        super(SVGPModel, self).__init__(variational_strategy)
        
        # Define kernel
        self.mean_module = gpytorch.means.ZeroMean()
        self.covar_module = gpytorch.kernels.RBFKernel()
    
    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)

# Select inducing points (e.g., 500 evenly spaced points)
M = 500
inducing_points = X[::N//M].unsqueeze(-1)  # Shape [M, 1]
model = SVGPModel(inducing_points)
likelihood = gpytorch.likelihoods.GaussianLikelihood()
```

```python
# Set the model and likelihood to training mode
model.train()
likelihood.train()

# Use the Adam optimizer
optimizer = torch.optim.Adam([
    {'params': model.parameters()},
    {'params': likelihood.parameters()},
], lr=0.01)

# Define the variational ELBO loss
mll = gpytorch.mlls.VariationalELBO(likelihood, model, num_data=N)

# Training parameters
epochs = 5
batch_size = 1024

# Create DataLoader for minibatches
from torch.utils.data import TensorDataset, DataLoader
dataset = TensorDataset(X.unsqueeze(-1), Y)
loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

for epoch in range(epochs):
    for batch_x, batch_y in loader:
        optimizer.zero_grad()
        output = model(batch_x)
        loss = -mll(output, batch_y)
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}/{epochs} - Loss: {loss.item():.3f}")
```

```python
# Switch to evaluation mode
model.eval()
likelihood.eval()

# Test inputs
X_test = torch.linspace(0, 10, 1000).unsqueeze(-1)
with torch.no_grad(), gpytorch.settings.fast_pred_var():
    predictions = likelihood(model(X_test))
    
# Plot results
mean = predictions.mean
lower, upper = predictions.confidence_region()

plt.figure(figsize=(10,6))
plt.scatter(X.numpy(), Y.numpy(), color='lightgrey', alpha=0.3, label='Training Data')
plt.plot(X_test.numpy(), mean.numpy(), color='blue', label='SVGP Mean')
plt.fill_between(X_test.squeeze().numpy(), lower.numpy(), upper.numpy(), color='blue', alpha=0.2, label='Confidence Interval')
plt.legend()
plt.title('Stochastic Variational GP Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Works well when the latent is smooth, without granual structure.

### Complex Latents

```python
torch.manual_seed(0)

# Generate complex data with multiple sinusoidal frequencies and a linear trend
N = 100000
X = torch.linspace(0, 10, N)

# Define the true underlying function
true_function = (
    0.5 * torch.sin(0.1 * X) +         # Very low-frequency component
    torch.sin(X) +                     # Standard frequency
    0.3 * torch.sin(10 * X) +          # Higher-frequency component
    0.2 * torch.sin(50 * X) +          # Very high-frequency component
    0.1 * X                            # Linear trend
)
```

```python
# Generate noisy observations
Y = true_function + torch.randn(N) * 0.2  # Adding noise

# Define the SVGP model and training as before
class SVGPModel(gpytorch.models.ApproximateGP):
    def __init__(self, inducing_points):
        variational_distribution = gpytorch.variational.CholeskyVariationalDistribution(inducing_points.size(0))
        variational_strategy = gpytorch.variational.VariationalStrategy(
            self, inducing_points, variational_distribution, learn_inducing_locations=True
        )
        super(SVGPModel, self).__init__(variational_strategy)
        
        self.mean_module = gpytorch.means.ZeroMean()
        self.covar_module = gpytorch.kernels.RBFKernel()
    
    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)

# Select inducing points
M = 500
inducing_points = X[::N//M].unsqueeze(-1)
model = SVGPModel(inducing_points)
likelihood = gpytorch.likelihoods.GaussianLikelihood()
```

```python
# Training setup
model.train()
likelihood.train()
optimizer = torch.optim.Adam([
    {'params': model.parameters()},
    {'params': likelihood.parameters()},
], lr=0.01)
mll = gpytorch.mlls.VariationalELBO(likelihood, model, num_data=N)

# Training loop with the modified dataset
epochs = 5
batch_size = 1024
dataset = TensorDataset(X.unsqueeze(-1), Y)
loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

for epoch in range(epochs):
    for batch_x, batch_y in loader:
        optimizer.zero_grad()
        output = model(batch_x)
        loss = -mll(output, batch_y)
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}/{epochs} - Loss: {loss.item():.3f}")

# Switch to evaluation mode
model.eval()
likelihood.eval()

# Testing and plotting
X_test = torch.linspace(0, 10, 1000).unsqueeze(-1)
with torch.no_grad(), gpytorch.settings.fast_pred_var():
    predictions = likelihood(model(X_test))

mean = predictions.mean
lower, upper = predictions.confidence_region()

# Calculate the true function values for the test points
true_function_test = (
    0.5 * torch.sin(0.1 * X_test.squeeze()) + 
    torch.sin(X_test.squeeze()) + 
    0.3 * torch.sin(10 * X_test.squeeze()) + 
    0.2 * torch.sin(50 * X_test.squeeze()) + 
    0.1 * X_test.squeeze()
)
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X.numpy(), Y.numpy(), color='lightgrey', alpha=0.3, label='Training Data')
plt.plot(X_test.numpy(), true_function_test.numpy(), color='green', linestyle='--', label='True Function')
plt.plot(X_test.numpy(), mean.numpy(), color='blue', label='SVGP Mean')
plt.fill_between(X_test.squeeze().numpy(), lower.numpy(), upper.numpy(), color='blue', alpha=0.2, label='Confidence Interval')
plt.legend()
plt.title('Stochastic Variational GP Regression with Complex Latent Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Cannot account for granular structures.
