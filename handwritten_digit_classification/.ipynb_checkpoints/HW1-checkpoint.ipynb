```python
import numpy as np

lambda_rate = 2  
time_interval = 10
mean_arrivals = lambda_rate * time_interval

E = np.random.poisson(mean_arrivals, 100)

mean_E = np.mean(E)

sample_variance_E = np.var(E, ddof=1)

# Output results
print(f"Mean number of arrivals: {mean_E}")
print(f"Sample variance: {sample_variance_E}")
```

```python
import matplotlib.pyplot as plt

plt.hist(E, bins=10, edgecolor='black')
plt.title('Histogram of Customer Arrivals (E1, E2, ..., E100)')
plt.xlabel('Number of Customer Arrivals')
plt.ylabel('Frequency')
plt.show()
```

```python
all_inter_arrival_times = []

for _ in range(100):
    current_time = 0
    while current_time < time_interval:
        inter_arrival_time = np.random.exponential(1 / lambda_rate)
        current_time += inter_arrival_time
        if current_time < time_interval:  
            all_inter_arrival_times.append(inter_arrival_time)

all_inter_arrival_times = np.array(all_inter_arrival_times)

num_long_intervals = np.sum(all_inter_arrival_times > 1)
percentage_long_intervals = (num_long_intervals / len(all_inter_arrival_times))

theoretical_value = np.exp(-2)

# Print results
print(f"Total number of inter-arrival times: {len(all_inter_arrival_times)}")
print(f"Percentage of inter-arrival times > 1 minute: {percentage_long_intervals:.4f}")
print(f"Theoretical value (e^-2): {theoretical_value:.4f}")
```
